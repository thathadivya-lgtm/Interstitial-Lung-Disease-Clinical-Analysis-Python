#### **Checking  and treating outliers:**

```python
# # Create a grid of 2 rows and 4 columns for displaying multiple boxplots
fig,axes=plt.subplots(2,4,figsize=(12,6))

# convert 2D array of axes into 1D array for easy access
axes=axes.flatten()

#Defining num columns:
num_col=['Age','Sex','BMI', 'FVC', 'FVC_pp', 'FEV1', 'FEV1_pp','FEV1_FVC_ratio']

# looping through each numerical column and create box plot
for i,col in enumerate(num_col):
    sns.boxplot(df[col],ax=axes[i])
    axes[i].set_title(col)

plt.tight_layout()
```

* The boxplots reveal the presence of outliers in BMI, FVC, FEV1, and FEV1/FVC ratio. Since these are **clinical measurements, the extreme values may represent genuine patient conditions rather than data entry errors**. Therefore, the **outliers were retained** and considered during the selection of appropriate imputation techniques.

* The **BMI column** contains 162 missing values. Since BMI is a numerical feature and the dataset does not provide separate height and weight measurements to recalculate BMI, the missing values will be imputed using the mean or median value.
* From the boxplot, the BMI boxplot shows the presence of outliers on both the lower and upper ends of the distribution.
* Since BMI contains missing values and outliers, **median imputation may be more robust than mean imputation** because the median is less affected by extreme values.

```python
# Imputing median value in 162 missing values of BMI
df['BMI']=df['BMI'].fillna(df['BMI'].median())
```

Missing values of pulmonary function measurements .png

* From the above screenshot,we can observe Spirometric_defect column and FVC, FVC_pp, FEV1, FEV1_pp, and FEV1_FVC_ratio columns also contains same rows of missing values.  
* The identical missing-value pattern across all spirometry-related columns suggests that lung function measurements were unavailable for these patients, while other clinical information was successfully recorded.
* Based on the above observations we should retain these missing rows.

From outlier analysis using boxplots revealed
* presence of extreme values in FVC, FEV1, and FEV1_FVC_ratio. Therefore, median imputation was chosen for these variables to reduce the influence of outliers.
* For FVC_pp and FEV1_pp, median imputation was used to ensure robustness against extreme values.

```python
# Imputing median value in FVC,FEV1 and FEV1_FVC_ratio
df['FVC']=df['FVC'].fillna(df['FVC'].median())
df['FEV1']=df['FEV1'].fillna(df['FEV1'].median())
df['FEV1_FVC_ratio']=df['FEV1_FVC_ratio'].fillna(df['FEV1_FVC_ratio'].median())
```

```python
# Imputing median value in FVC_pp,FEV1_pp
df['FVC_pp']=df['FVC_pp'].fillna(df['FVC_pp'].median())
df['FEV1_pp']=df['FEV1_pp'].fillna(df['FEV1_pp'].median())
```

* Since spirometric defect is a categorical variable and its classification depends on clinical interpretation of lung function tests, missing values were imputed using the most frequent category(mode) rather than deriving new classifications from imputed numerical values.

```python
# Imputing mode value in spirometric_defect column mising values
df['Spirometric_defect']=df['Spirometric_defect'].fillna(df['Spirometric_defect'].mode()[0])
```

* The identical missing-value pattern in Invasive_procedure and Diagnostic_value suggests that diagnostic procedure information was unavailable for the same group of patients.
* Since Invasive_procedure and Diagnostic_value columns represents categorical information and there is no reliable method exists to infer the procedure performed for individual patients, the missing values were imputed using the mode.



```python
# Imputing mode value in  Invasive_procedure and Diagnostic_value column mising values
df['Invasive_procedure']=df['Invasive_procedure'].fillna(df['Invasive_procedure'].mode()[0])
df['Diagnostic_value']=df['Diagnostic_value'].fillna(df['Diagnostic_value'].mode()[0])
```

* The Stage_sarcoidosis column contains missing values because sarcoidosis staging is only applicable to patients diagnosed with Sarcoidosis.
* For patients with other ILD diagnoses, this information is not relevant.
* Therefore, the missing values were replaced with "Not Applicable" instead of using mode imputation.

```python
# Filling missing values of Stage_sarcoidosis with "Not Applicable"
df['Stage_sarcoidosis']=df['Stage_sarcoidosis'].fillna('Not_Applicable')
```

```python
# Rechecking if all the missing values are filled or not
df.isnull().sum()
```
